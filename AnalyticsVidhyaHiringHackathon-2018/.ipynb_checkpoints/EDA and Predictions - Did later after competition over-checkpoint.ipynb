```python
def draw_normality_curves(series):
    fig, ax =plt.subplots(figsize=(16,5), ncols=3, nrows=1, squeeze=False)
    sm.qqplot(series, line='45',ax=ax[0][0])
    sns.distplot(series, kde_kws=dict(bw=0.5), hist_kws= dict(edgecolor="k"),ax=ax[0][1])
    sns.boxplot(series,ax=ax[0,2])
```

```python
draw_normality_curves(data.temperature)

# data.temperature1 = 1/data.temperature

# draw_normality_curves(data.temperature1)
```

```python
draw_normality_curves(data.var1)

# data.var11 = np.sqrt(data.var1)

# draw_normality_curves(data.var11)
```

```python
draw_normality_curves(data.pressure)

data.pressure1 = np.log(data.pressure)

draw_normality_curves(data.pressure1)
```

```python
draw_normality_curves(data.windspeed)

data.windspeed1 = np.sqrt(np.log(data.windspeed))

draw_normality_curves(data.windspeed1)
```

```python
draw_normality_curves(data.electricity_consumption)
data.electricity_consumption1 = np.sqrt(np.sqrt(data.electricity_consumption))
draw_normality_curves(data.electricity_consumption1)
```

```python
data.electricity_consumption=data.electricity_consumption1
data.windspeed=data.windspeed1
```

```python
data.datetime = pd.to_datetime(data.datetime)

#data["date"] = [d.day for d in data["datetime"]]
data["month"] = [d.month for d in data["datetime"]]
data["year"] = [d.year for d in data["datetime"]]
data["hour"] = [d.hour for d in data["datetime"]]
data["hour"] = [d.hour for d in data["datetime"]]
data['weekday']= [d.weekday() for d in data["datetime"]]

data_fet_eng_1 = data
del data_fet_eng_1["datetime"]
del data_fet_eng_1["ID"]

var2_dummy = pd.get_dummies(data_fet_eng_1.var2)
del var2_dummy['C']
data_fet_eng_2 = pd.concat([data_fet_eng_1,var2_dummy],axis=1)
del data_fet_eng_2['var2']
y_train = data_fet_eng_2['electricity_consumption']
del data_fet_eng_2['electricity_consumption']
#data_fet_eng_2.to_csv('train_eng.csv')
data_fet_eng_2.head()
```

```python
test = pd.read_csv('test.csv')
test_ids = test['ID']
test.datetime = pd.to_datetime(test.datetime)
#test["date"] = [d.day for d in test["datetime"]]
test["month"] = [d.month for d in test["datetime"]]
test["year"] = [d.year for d in test["datetime"]]
test["hour"] = [d.hour for d in test["datetime"]]
test["hour"] = [d.hour for d in test["datetime"]]
test['weekday']= [d.weekday() for d in test["datetime"]]
test.windspeed = np.sqrt(np.log(test.windspeed))
test.windspeed = np.sqrt(np.log(test.windspeed))

var2_dummy = pd.get_dummies(test.var2)
del var2_dummy['C']
test = pd.concat([test,var2_dummy],axis=1)

del test['var2']
del test["datetime"]
#test.to_csv('test_fet.csv')
del test["ID"]
test.head()
```

```python
import xgboost
xgb = xgboost.XGBRegressor(n_estimators=370, base_score=0.5,learning_rate=0.08, gamma=0, subsample=0.9,
                           colsample_bytree=1, max_depth=5)

xgb.fit(data_fet_eng_2,y_train)
```

```python
predictions = xgb.predict(test)
predictions = np.square(np.square(predictions)).astype('int32')

sub = pd.DataFrame({'ID': test_ids, 'electricity_consumption': predictions})
filename = 'sub_returns.csv'
sub.to_csv(filename, index=False)
```
